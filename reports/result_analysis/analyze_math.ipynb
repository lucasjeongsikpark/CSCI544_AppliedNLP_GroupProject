```python
# import math_result.xlsx into df
import pandas as pd
df = pd.read_excel('math_result.xlsx')
df['human'] = 0
for idx in range(len(df)):
    human1 = df.at[idx, 'human1 (Manav)']
    human2 = df.at[idx, 'human2 (Pranshu)']
    human1_score = list(map(int, human1.split(',')))
    human2_score = list(map(int, human2.split(',')))
    llama_score = human1_score[0] + human2_score[0]
    distill_score = human1_score[1] + human2_score[1]
    if llama_score > distill_score:
        df['human'][idx] = 0
    elif llama_score < distill_score:
        df['human'][idx] = 1
    else:
        df['human'][idx] = 5
df = df.drop(columns=['human1 (Manav)', 'human2 (Pranshu)'])
df
```

```python
# col_names = Debatable-Intelligence	MORE	DEBATE	ChatEval	Single Agent	

col_names = ['Debatable-Intelligence', 'MORE', 'DEBATE', 'ChatEval', 'Single Agent']
for i in range(len(df)):
    for col_name in col_names:
        scores_str = df.at[i, col_name]
        scores_list = list(map(int, scores_str.split(',')))
        if scores_list[0] > scores_list[1]:
            flag = 0
        elif scores_list[0] < scores_list[1]:
            flag = 1
        else:
            flag = 5
        df.at[i, col_name] = flag
```

```python
for i in range(4):
    score = [0,0,0,0,0]
    for j in range(i*30, i*30+30, 1):
        for idx, col_name in enumerate(col_names):
            if df.at[j, col_name] == df.at[j, 'human']:
                score[idx] += 1
    print(score)
```
